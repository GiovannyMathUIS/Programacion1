```python
%%writefile promedio.py
import statistics
def calcular_promedio(n1, n2, n3):
  return (n1 + n2 + n3) / 3

def mediana(n1, n2, n3):
  return statistics.median([n1, n2, n3])
def moda(n1, n2, n3):
  return statistics.mode([n1, n2, n3])
```
