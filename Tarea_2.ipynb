```python
def nacci(manolo):
    if manolo <= 0:
        return []

    a, b = 0, 1
    secuencia = []

    while b < manolo:
        secuencia.append(b)
        a, b = b, a + b

    return secuencia

try:
    n = int(input("Ingrese el límite: "))
    print("Secuencia:", nacci(n))
except ValueError:
    print("Debes ingresar un número válido")
```

```python
numeros = list(range(777))

pares = [n for n in numeros if n % 2 == 0]
impares = [n for n in numeros if n % 2 != 0]

print("Pares:", pares)
print("Impares:", impares)
print("Suma total:", sum(numeros))
```

```python
def pepe(x):
    if x % 3 == 0:
        return "Múltiplo de 3"
    return f"No es múltiplo, residuo: {x % 3}"


try:
    x = int(input("Número: "))
    print(pepe(x))
except ValueError:
    print("Entrada inválida")
```

```python
def clasificar(x):
    if x < 0:
        return "Negativo"
    elif x == 0:
        return "Cero"
    elif x == 1:
        return "Uno"
    elif 2 <= x <= 10:
        return "meh"
    else:
        return "perfect"


x = int(input("Número: "))
print(clasificar(x))
```
