```python
a = {6, 2, 3, 4, 5}
b = {1, 2, 3, 4}
```

```python
a.difference(b) #analiza los dos conjuntos y devuelve un conjunto con todos lo valores que estan en el conjunto a pero no en el b
#en este caso devuelve {5, 6}
```

```python
a.symmetric_difference(b) # El metodo a.symmetric_difference(b) muestra todos los valores que se encuentran en uno solo de los copnjuntos, osea los que no estan en a y b al mismo tiempo.
# en este caso devuelve {1, 5, 6}
```

```python
c = {1, 2, 3, 4, 5}
d = {1, 2, 3}
```

```python
c.issuperset(d)# verifica si todos los valores del conjunto d se encuentran en c, si este es el caso imprime true. si por el contrario no todos los valores de d se encuentran en c imprime false
# en este caso imprime true
```

```python
c.isdisjoint(d)# verifica si todos los valores del conjunto d se encuentran en c, si este es el caso imprime false.
# en este caso imprime false
```
